```python
import re
```

```python
file_path = "text_for_plot.txt"  # Path to the input text file
```

```python
train_loss = []
valid_loss = []
with open(file_path, "r") as file:
    for line in file:
        line = line.strip()  # Remove leading/trailing whitespaces
        if line.startswith("100%"):
            if "train_loss" in line:
                train_loss_value = re.search(r"train_loss=([\d.]+)", line)
                if train_loss_value:
                    train_loss.append(float(train_loss_value.group(1)))

            elif "valid_loss" in line:
                valid_loss_value = re.search(r"valid_loss=([\d.]+)", line)
                if valid_loss_value:
                    valid_loss.append(float(valid_loss_value.group(1)))
```

```python
import matplotlib.pyplot as plt
epochs = list(range(1, len(train_loss) + 1))  # Assuming 'train_loss' and 'valid_loss' are arrays of loss values
plt.plot(epochs, train_loss, label='Train Loss')
plt.plot(epochs, valid_loss, label='Valid Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Train and Valid Loss vs Epochs')
plt.legend()
plt.show()
```
