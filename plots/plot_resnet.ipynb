```python
resnet_output = '/scratch/sg7729/DL_project/SLIP_Github/baseline classifier/classifier output/ResNet18_pokemon_ouptut'
```

```python
train_acc = []
val_acc = []
val_loss = []
train_loss = []

with open(resnet_output, 'r') as file:
    lines = file.readlines()
    for line in lines:
        if line.startswith('[Epoch'):
            # Extract train loss and accuracy
            train_info = line.split('Train Loss: ')
            if len(train_info) >= 2:
                train_info = train_info[1].split(', Train Acc: ')
                train_loss.append(float(train_info[0]))
                train_acc.append(float(train_info[1].rstrip('%\n')))
            else:
                # Extract val loss and accuracy
                val_info = line.split('Val Loss: ')[1].split(', Val Acc: ')
                val_loss.append(float(val_info[0]))
                val_acc.append(float(val_info[1].rstrip('%\n')))

```

```python
import matplotlib.pyplot as plt

# Plotting the train loss 
# Epoch 35 achieved best validation loss, subsequent epochs might ovrefit.
plt.plot(range(1, 35 + 1), train_loss[:35], label='Train Loss')

# Adding labels and title to the plot
plt.xlabel('Epoch')
plt.ylabel('Baseline Classifier Loss')
plt.title('Train Loss vs Validation Loss')
plt.legend()

# Displaying the plot
plt.show()
```
